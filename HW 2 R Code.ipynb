```r
#1.5: Plotting the log-likelihood l(lambda) in R for a chosen sample:

#A chosen sample being stored as well as the provided alpha and theta values, and the mean:
alpha = 2
theta = 3

x = c(4, 8, 5, 9, 10, 4, 4, 4)
mean.x = 3*alpha/(alpha - 1)
print(mean.x)
print(mean(x))

alpha.seq = seq(.1, 20, by = 0.1)
print(length(alpha.seq))
l = numeric()
print(l)

for(i in 1:length(alpha.seq))
{
  l[i] = length(x)*log(2) + length(x)*2*log(3) - (2+1)*sum(log(x))
}

#par(mar = c(5,5,5,5), oma = c(5,5,5,5)) #Changing the plot size

#Plotting the log-likelihood
plot(alpha.seq,l,xlab = 'alpha', ylab = 'log-likelihood', type = 'p', main = 'MLE')
abline(v = length(x)/(sum(log(x)) - (length(x)*log(3))), col = 2)
```
